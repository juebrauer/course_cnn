# Collect result data from training history files

```python
result_folder = "exp_results_learnrate"
```

```python
import os
from os import listdir
from os.path import isdir, isfile, join
```

```python
filenames = [result_folder + "/" + f
             for f in listdir(result_folder)
                 if isfile(join(result_folder, f)) and f.endswith(".history")]
```

```python
filenames[:10]
```

```python
import pickle
import matplotlib.pyplot as plt

all_exp_results = []

learn_rates_analyzed = []

for filename in filenames:
    
    # 1. Read in the training/testing history from
    #    the .history file using pickle    
    fhandle = open(filename, "rb")
    history = pickle.load(fhandle)
    #print("history is of type {0}".format(type(history)))
    #print("history: {0}".format(history))
    fhandle.close()
    
    
    # 2. Get the training and testing classification rates
    #    Get learn rate
    learn_rate = history["learn_rate"]
    tr_cl_rates = history["cl_rate_train"]
    tst_cl_rates = history["cl_rate_test"]
    final_tst_cl_rate = tst_cl_rates[-1]
    
    
    # 3. Show training / test classification rates as a graph?
    if True:
        plt.plot(tr_cl_rates,  color="red",   marker="*", label="train acc")
        plt.plot(tst_cl_rates, color="green", marker="+", label="test acc")
        plt.legend()
        plt.grid()
        plt.ylim(0.0, 1.0)
        plt.title("{0}\nlearn_rate: {1}".format(filename, learn_rate))
        plt.show()
            
    
    # 4. show some info
    nr_epochs = len(tr_cl_rates)
    print("Learn rate {0:.5f} --> final test classification rate after {1} epochs: {2:.2f}\n\n"
          .format(learn_rate, nr_epochs, final_tst_cl_rate))
    
    
    # 5. Compile a list of all learning rates that we analyzed in the experiments
    if not learn_rate in learn_rates_analyzed:
        learn_rates_analyzed.append( learn_rate )
            
   
    # 6. Compile a dictionary with the results for this experiment and store it in
    #    the list of all experiment results
    d = {"final_tst_cl_rate" : final_tst_cl_rate,
         "learn_rate" : learn_rate }
    if len(all_exp_results)<10 and False:
        print("\nData from {0}".format(filename))
        print(d)
    all_exp_results.append( d )
        
    
    
nr_experiments_conducted = len(all_exp_results)
print("\nI have collected results from {0} experiments".format(nr_experiments_conducted))

learn_rates_analyzed.sort()
print("\nThe following learn rates have been analyzed in your experiments: {0}"
      .format(learn_rates_analyzed))
    
```

# Plot the results

```python
import numpy as np

xt = np.arange(1,len(learn_rates_analyzed))

plt.xticks(xt, learn_rates_analyzed)

sum_working_models = 0.0
nr_working_models = 0
for exp_dict in all_exp_results:
         
    learn_rate = exp_dict["learn_rate"]
    x = learn_rates_analyzed.index(learn_rate)
    
    y = exp_dict["final_tst_cl_rate"]    
        
    plt.plot(x,y, marker=".", color="black")

    if y > 0.15:
        nr_working_models +=1
        sum_working_models += y            

avg_clf_rate = sum_working_models/nr_working_models
print("Average test classification rate: {0:0.2f}"
      .format(avg_clf_rate))

plt.grid()
plt.show()
```

# Conclusions

We can see that some learn rates (e.g. 0.1, 1.0, 10.0) did no work at all.

Some learn rates worked, but the training probably needs more time to converge (e.g. 1e-05, 1e-04, 1e-03).

The learn rate 0.01 was the best in this case.
