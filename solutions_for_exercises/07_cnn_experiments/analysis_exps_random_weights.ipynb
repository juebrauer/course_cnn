# Analysis of the experiment results

```python
result_folder = "exp_results_rnd_conv_weights/tmp_results"
import os
from os import listdir
from os.path import isdir, isfile, join
filenames = [result_folder + "/" + f
             for f in listdir(result_folder)
                 if isfile(join(result_folder, f)) and f.endswith(".history")]
filenames[:10]
```

```python
import pickle
import matplotlib.pyplot as plt

all_exp_results = []

for filename in filenames:
    
    # 1. Read in the training/testing history from
    #    the .history file using pickle    
    fhandle = open(filename, "rb")
    history = pickle.load(fhandle)
    #print("history is of type {0}".format(type(history)))
    #print("history: {0}".format(history))
    fhandle.close()
            
    
    # 2. Get the training and testing classification rates    
    tr_cl_rates = history["cl_rate_train"]
    tst_cl_rates = history["cl_rate_test"]
    final_tst_cl_rate = tst_cl_rates[-1]
    
    
    # 3. Which dataset was used in this experiment?
    #    Did we use random CONV layer weights in this experiment?
    dataset_name = history["dataset_name"]
    rnd_conv_weights  = history["use_rnd_weights"]    
    
    
    # 4. Show training / test classification rates as a graph?
    if True:
        plt.plot(tr_cl_rates,  color="red",   marker="*", label="train acc")
        plt.plot(tst_cl_rates, color="green", marker="+", label="test acc")
        plt.legend()
        plt.grid()
        plt.ylim(0.0, 1.0)
        plt.title("{0}\nrnd conv weights: {1}\ndataset: {2}"
                  .format(filename, rnd_conv_weights, dataset_name))
        plt.show()
            
    
    # 5. show some info
    nr_epochs = len(tr_cl_rates)
    print("Final test classification rate after {0} epochs: {1:.2f}\n\n"
          .format(nr_epochs, final_tst_cl_rate))
            
   
    # 6. Compile a dictionary with the results for this experiment and store it in
    #    the list of all experiment results
    d = {"dataset_name" : dataset_name,
         "rnd_conv_weights" : rnd_conv_weights,
         "final_tst_cl_rate": final_tst_cl_rate}
    if len(all_exp_results)<10 and False:
        print("\nData from {0}".format(filename))
        print(d)
    all_exp_results.append( d )
        
    
    
nr_experiments_conducted = len(all_exp_results)
print("\nI have collected results from {0} experiments".format(nr_experiments_conducted))
```

```python
import numpy as np

labels = ["rnd conv weight", "conv weights trained"]

xt = np.arange(1,len(labels)+1)

plt.xticks(xt, labels)

sum_test_cl_rates = 0.0
N = 0

l1_rnd_conv_weights_acc = []
l2_trained_conv_weights_acc = []

for d in all_exp_results:
         
    if d["dataset_name"]!="imagenette":
        continue
        
    y = d["final_tst_cl_rate"]    
    
    if d["rnd_conv_weights"]:
        x=1
        l1_rnd_conv_weights_acc.append( y )
    else:
        x=2
        l2_trained_conv_weights_acc.append( y )
        
    plt.plot(x,y, marker=".", color="black")
    
    sum_test_cl_rates += y
    N +=1
    
avg_clf_rate = sum_test_cl_rates / N
print("Average test classification rate: {0:0.2f}"
      .format(avg_clf_rate))

plt.grid()
plt.show()
```
